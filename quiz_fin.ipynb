```python

# The Python quiz game

print("Hai ... Welcome to Python quiz!")

print('Enter your name:  ')

name = input()

print("Get Ready "+ name)

questions = ("1. How do you start writing a while loop in Python? ",
                       "2. Which statement is used to stop a loop? ",
                       "3. Which method can be used to replace parts of a string? ",
                       "4. Which operator is used to multiply numbers? ",
                       "5. Which method can be used to remove any whitespace from both the beginning and the end of a string?")

options = (("A. x > y while {", "B. while x > y {", "C. while x > y:", "D. while (x > y)"),
           ("A. return", "B. stop", "C. exit", "D. break"),
           ("A. replaceString()  ", "B. replace()  ", "C. switch()", "D. repl()"),
           ("A. &","B. $", "C.*", "D. %"),
           ("A. strip()", "B. len()", "C. ptrim()", "D. trim()"))

answers = ("A", "D", "B", "C", "A")
guesses = []
score = 0
question_num = 0

for question in questions:
    print("----------------------")
    print(question)
    for option in options[question_num]:
        print(option)

    guess = input("Enter (A, B, C, D): ").upper()
    guesses.append(guess)
    if guess == answers[question_num]:
        score += 1
        print("CORRECT!")
    else:
        print("INCORRECT!")
        print(f"{answers[question_num]} is the correct answer")
    question_num += 1

print("-----------------------------")     
print("       RESULTS        ") 
print("------------------------------") 


print("answers: ", end="")
for answer in answers:
    print(answer, end=" ")
print()

score = int(score / len(questions) * 100)
print("Python quiz score for " + name + f" is: {score}%")

      
```
